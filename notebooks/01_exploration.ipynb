```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
```

```python
# 1. Filter out negative and zero fare amounts
# We assume a valid trip must have a positive fare
df_cleaned = df[df['fare_amount'] > 0].copy()

# 2. Filter trip distance
# Remove trips with 0 distance or unrealistic distances (e.g., > 100 miles)
df_cleaned = df_cleaned[(df_cleaned['trip_distance'] > 0) & (df_cleaned['trip_distance'] < 100)]

# 3. Filter passenger count
# Standard taxis usually carry 1 to 6 passengers
df_cleaned = df_cleaned[(df_cleaned['passenger_count'] > 0) & (df_cleaned['passenger_count'] < 7)]

# 4. Remove rows with missing values (NaNs) in critical columns
# This ensures our analysis is based on complete records
df_cleaned = df_cleaned.dropna(subset=['passenger_count', 'RatecodeID', 'congestion_surcharge'])

# 5. Check the new shape and statistics of the cleaned data
display(
    "Original Shape:", df.shape,
    "Cleaned Shape:", df_cleaned.shape,
    "Data Lost (%):", round((1 - (df_cleaned.shape[0] / df.shape[0])) * 100, 2)
)

# Show updated descriptive statistics
display("Cleaned Data Description:", df_cleaned.describe().round(2).T)
```

```python
# 1. Filter RatecodeID to include only official NYC codes (1-6)
df_cleaned = df_cleaned[df_cleaned['RatecodeID'] <= 6]

# 2. Dynamic Outlier Removal using 99th Percentile
# Instead of using fixed numbers like $500 or 20 miles, 
# we calculate the threshold that covers 99% of the data.
dist_threshold = df_cleaned['trip_distance'].quantile(0.99)
fare_threshold = df_cleaned['total_amount'].quantile(0.99)

df_cleaned = df_cleaned[
    (df_cleaned['trip_distance'] <= dist_threshold) & 
    (df_cleaned['total_amount'] <= fare_threshold)
]

# 3. Handle 'extra' column negatives
df_cleaned = df_cleaned[df_cleaned['extra'] >= 0]

print("Data Polishing Complete (using 99th Percentile).")
print(f"New Records Count: {len(df_cleaned)}")
print(f"Threshold for Distance: {dist_threshold:.2f} miles")
print(f"Threshold for Fare: ${fare_threshold:.2f}")
print(f"New Maximum Fare: ${df_cleaned['total_amount'].max():.2f}")
```
